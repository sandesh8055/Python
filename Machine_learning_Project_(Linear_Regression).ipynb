Step 1: Gathering Data

```python
#import packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

```python
data.info()
```

Step 2 : Data Preprocessing

```python
#One Hot Encoding - It will convert categorical variables into numerical variables.
data['sex']=data['sex'].apply({'female':0,'male':1}.get)
```

```python
data.head()
```

```python
data["smoker"]=data["smoker"].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':0,'northeast':2,'northwest':3}.get)
```

```python
data.head()
```

Step 3: Dividing the data into dependent and independent variables.

```python
x=data[['age','sex','bmi','children','smoker','region']]  #independent data
y=data['expenses']  #dependent data
```

Step 4: Splitting the data into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#test_size=0.2 means that 20% of the data will be used to test the accuracy of the model
#80% will be used to train the model
```

Step 5: Creating a ML model using linear regression algorithm

```python
#import linear regression algorithm
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression() #We are activating the algorithm for use
```

```python
#training the model using .fit()
linear_model.fit(x_train,y_train)
```

```python
y_pred=linear_model.predict(x_test) #.predict used for predicting the future values
```

```python
print(y_pred) #y_pred is the amout of insurance calculated by the ML model
```

```python
#checking the accuracy of my ML model
accuracy=linear_model.score(x,y)
```

```python
print("Accuracy of my ML model is", accuracy)
```

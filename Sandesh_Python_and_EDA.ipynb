a. Left Skewness(-vely Skewed) : Mean is low , Median is high, left side graph is more close to the x-axis.

Hypothesis Testing:

1. H0 - Null Hypothesis, We try to reject. It is something that is involved in context/problem statement.
2. H1 - Alternate Hypothesis, opposite of H0.

eg. there is a straw manufacturing company who claims that average radius of straws manufactured is 0.05 mm. what is H0 and H1.

H0: average radius is 0.05mm.
H1: average radius is not 0.05mm.
so for H1, average radius can be < 0.05 mm.
        or average radius can be > 0.05 mm.  (So it is a two tailed test)

2. there is a straw manufacturing company who claims that average radius of straws manufactured is atleast 0.05 mm. what is H0 and H1.

H0= average radius is => 0.05 mm.
H1= average radius is less than 0.05 mm. (its a left one tailed test)

3. there is a pipe manufacturing company who claims that average radius of pipe manufactured is atmax 0.05 mm. what is H0 and H1.

H0 : average radius is <= 0.05mm
H1 : average radius is > 0.05 mm (right One tailed test)

level of significance = 100 - level of confidance %
